```r
virtual_shovel <- bowl |> rep_sample_n(size = 50, reps = 1000) |>
  mutate(is_red = (color =='red')) |>
  summarize(num_red = sum(is_red)) |>
  mutate(prop_red = num_red / 50)
```

```r
ggplot(data = virtual_shovel, mapping = aes(x = prop_red)) +
  geom_histogram()
```
